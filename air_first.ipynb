```python
%matplotlib inline
import os
import seaborn as sns
from pycaret.regression import *
from catboost import CatBoostRegressor
import matplotlib.pyplot  as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,StandardScaler,MinMaxScaler
from sklearn.model_selection import KFold,cross_val_score
import math
import numpy as np
import pandas as pd
from datetime import datetime
sns.set()
df = pd.read_csv(r'C:/Users/USER/kagglesync/Zindi_Air/Train.csv')
df_test = pd.read_csv(r'C:/Users/USER/kagglesync/Zindi_Air/Test.csv')
df_sub = pd.read_csv(r'C:/Users/USER/kagglesync/Zindi_Air/SampleSubmission.csv')

cols_to_fillna = ['humidity','temp']

for i in cols_to_fillna:
    df[i].fillna(np.mean(df[i]),inplace=True)

print(df.info())
labelencoder=LabelEncoder()
df['site']=labelencoder.fit_transform(df['site'])

df['created_at'] = df['created_at'].apply(lambda x:datetime.strptime(x,"%Y-%m-%d %H:%M:%S+03:00"))
df['created_at']=labelencoder.fit_transform(df['created_at'])
cols_to_drop = ['ID','landform_90m','long']

df.drop(cols_to_drop,axis=1,inplace=True)
##verify that the train data has been preprocessed
print(df.info())

##visualize for collineariity via heatmap
sns.heatmap(df.corr())

##normalize the dataset
minmax_scale = MinMaxScaler()

X_1 = df.drop(['ref_pm2_5'],axis=1)
data_x = pd.DataFrame(minmax_scale.fit_transform(X_1),columns=X_1.columns)
data_x['ref_pm2_5'] = df['ref_pm2_5']
print(data_x.head())
Y = data_x['ref_pm2_5']
X = data_x.drop(['ref_pm2_5'],axis=1)

print('sliceddd')
print(X.head())
print('slicced 2')
print(Y.head())

##preprocess the test data
df_test['site']=labelencoder.fit_transform(df_test['site'])

for i in cols_to_fillna:
    df_test[i].fillna(np.mean(df_test[i]),inplace=True)

df_test['created_at'] = df_test['created_at'].apply(lambda x:datetime.strptime(x,"%Y-%m-%d %H:%M:%S+03:00"))
df_test['created_at']=labelencoder.fit_transform(df_test['created_at'])

df_test.drop(cols_to_drop,axis=1,inplace=True)

df_test_frame = pd.DataFrame(minmax_scale.fit_transform(df_test),columns=df_test.columns)

##verify that the test data has been preprocessed
print(df_test_frame.info())

cat_reg = CatBoostRegressor(random_state=10,eval_metric='RMSE',
                            boosting_type='Plain',boost_from_average=True)

# cv = KFold(n_splits=5, random_state=42, shuffle=True)
##perform kfold cv achieved a mean rmse value of 12.52319030157003 on validation


# print('preparing to split')
# for train_index,test_index in cv.split(X):
#     print('split started')
#     print(train_index,test_index)
#     X_train,X_test,Y_train,Y_test = X.iloc[train_index],X.iloc[test_index],Y.iloc[train_index],Y.iloc[test_index]
#     cat_reg.fit(X_train, Y_train)
#     cross_val = cross_val_score(cat_reg,X,Y,scoring='neg_root_mean_squared_error',cv=5)
#     print(cross_val)


cat_reg.fit(X,Y)



pred = cat_reg.predict(df_test_frame)
print(pred[:5])
df_sub['ref_pm2_5'] = pred

df_sub.to_csv(r'C:/Users/USER/kagglesync/Zindi_Air/third.csv',index=False)
```
